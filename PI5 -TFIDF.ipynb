```python
import pandas as pd
import nltk
stopwords = nltk.corpus.stopwords.words('portuguese')
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
df = pd.read_csv('/kaggle/input/reviews-completas/Reviews Completo.csv',encoding='utf-8')
df
```

```python
df2 = df.copy()
for i in range(0,len(df2)):
    
    text = df.iloc[i,3]
    df2.iloc[i,3] = ''.join(c for c in text if not c.isdigit())
```

```python
df2
```

```python
vectorizer = CountVectorizer(stop_words=stopwords)
vetores = vectorizer.fit_transform(df2.iloc[:,3])
vocab = vectorizer.get_feature_names_out()
print(vocab)
```

```python
dfC = pd.DataFrame(data = vetores.toarray(),columns=vocab)
dfC
```

```python
vectoTf = TfidfVectorizer(stop_words=stopwords)
vectors = vectoTf.fit_transform(df2.iloc[:,3])
df_Tf = pd.DataFrame(data=vectors.toarray(),columns = vocab)
```
